#### Importing

```python
import pandas as pd
import numpy as np
```

#### Datasets

```python
tourism_df = pd.read_csv("Tourism_Hospitality_Industry_Analysis.csv")
policy_df = pd.read_csv("Tourism_City_Policy_Data.csv")
```

#### Set indexes for the JOIN using Country and City as compound keys

```python
tourism_df_indexed = tourism_df.set_index(['Country', 'City'])
policy_df_indexed = policy_df.set_index(['Country', 'City'])
```

#### Left Join

```python
left_join_df = tourism_df_indexed.join(
    policy_df_indexed, 
    how="left", 
    lsuffix='_tourism', 
    rsuffix='_policy'
)
```

```python
print(left_join_df.isna().sum())
```

#### I chose a left merge becasue my main dataset is the tourism dataset and I want to keep all of its rows regardless of where there is a matching policy data. 
#### By merging on the Country and City columns, all rows from the main tourism dataset are retained, while policy data is added where available. Only the 10 matching pairs from thew policy dataset are merged.
#### NA values appear in the 5 new policy columns for the 397 unmatched tourism records. These represent cities with no policy data.
